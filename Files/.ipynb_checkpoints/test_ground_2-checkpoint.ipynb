```python
import pandas as pd
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import SGDClassifier
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import seaborn as sns

import nltk
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer 
from wordcloud import WordCloud

from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, cross_val_predict
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, roc_curve, precision_recall_curve, classification_report
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
```

```python
df = pd.read_csv("../Data/disagreement_data.csv")
```

```python
df_2 = pd.read_csv("../Data/disagreement_data_with_appeal.csv")
```

```python
len(df)
```

```python
len(df_2)
```

```python
df.head()
```

```python
df = df.sample(frac=1)
```

```python
df.head()
```

```python
X = df['text']
y = df['label']
```

```python
lemmatizer = WordNetLemmatizer()

X_lemmatized = []

for document in X:
    word_list = nltk.word_tokenize(document)
    lemmatized_output = ' '.join([lemmatizer.lemmatize(w) for w in word_list])
    X_lemmatized.append(lemmatized_output)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_lemmatized, y, test_size=0.2, random_state=None)
```

```python
#count_vect = CountVectorizer(lowercase=True, stop_words='english', ngram_range=(1, 2))
#count_vect = CountVectorizer(lowercase=True, stop_words='english')
#count_vect = CountVectorizer(lowercase=True, stop_words='english', binary=True)
count_vect = CountVectorizer()


X_train_counts = count_vect.fit_transform(X_train)
print(X_train_counts.shape)

print("Type of the occurance count matrix (should be sparse): ")
print(type(X_train_counts))

X_test_counts = count_vect.transform(X_test)
print(X_test_counts.shape)

tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
print(X_train_tfidf.shape)

X_test_tfidf = tfidf_transformer.transform(X_test_counts)
print(X_test_tfidf.shape)
```

## SVM

```python
param_grid = [
  {'C': [1, 10, 100, 1000], 'kernel': ('linear', 'poly', 'rbf'), 'tol': [0.1, 0.01, 0.001]},
 ]
```

```python
svc = svm.SVC()
clf = GridSearchCV(svc, param_grid, scoring='accuracy', cv=10)
clf.fit(X_train_counts,y_train)
```

```python
clf.best_params_
```

```python
from sklearn import svm

svc = svm.SVC(gamma='auto', kernel='rbf', tol=0.01, C=100)
svc.fit(X_train_counts,y_train)
svc_prediction = svc.predict(X_test_counts)
accuracy_score(svc_prediction, y_test)
```

```python
svc = svm.SVC()
clf = GridSearchCV(svc, param_grid, scoring='accuracy', cv=10)
clf.fit(X_train_tfidf,y_train)
```

```python
clf.best_params_
```

## Random Forest

```python
param_grid = [
  {'max_features' : ['auto', 'sqrt'], 'max_depth' : [int(x) for x in np.linspace(10, 110, num = 11)], 'min_samples_split' : [2, 5, 10]
, 'min_samples_leaf' : [1, 2, 4]},
 ]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
clf = GridSearchCV(rf, param_grid, scoring='accuracy', cv=10)
clf.fit(X_train_counts,y_train)
```

```python
clf.best_params_
```

```python
rf = RandomForestClassifier(max_depth= 50, max_features= 'sqrt', min_samples_leaf=1,min_samples_split= 10)

rf.fit(X_train_counts,y_train)
rf_prediction = rf.predict(X_test_counts)
accuracy_score(rf_prediction, y_test)
```
