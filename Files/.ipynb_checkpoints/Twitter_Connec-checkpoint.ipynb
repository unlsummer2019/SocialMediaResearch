```python
class MyStreamer(TwythonStreamer):
    def on_success(self, data):
        count = 0
        t = []
        if 'text' in data:
            print(data['text'])
            k.append(data['text'])
            count = count + 1
            print(count)
            

    def on_error(self, status_code, data):
        print(status_code)
        self.disconnect()
```
