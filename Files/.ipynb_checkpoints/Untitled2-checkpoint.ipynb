```python
def decrease1(b):
    a = []
    for i in range(len(b)):
        if len(a) == 0:
            a.append(b[i])
        else:
            count = 0
            for j in range(len(a)):
                if a[j]['text'] != b[i]['text']:
                    count = count + 1
            
                if count == len(a):
                    a.append(b[i])
    
    return a

def decrease2(b):
    a = []
    for i in range(len(b['statuses'])):
        if len(a) == 0:
            a.append(b['statuses'][i]['text'])
        else:
            count = 0
            for j in range(len(a)):
                if a[j] != b['statuses'][i]['text']:
                    count = count + 1
            
                if count == len(a):
                    a.append(b['statuses'][i]['text'])
    
    return a
```
